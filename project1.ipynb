# make a directory can add, remove, find, show contact 

```python
lienlac = { \n    'nguyễn văn a': '0986765421', 
    'trần văn b' : '0254345275', 
    'toni kroos' : '0365674858' 
}
while True : 
    print('trinh quan ly danh ba') 
    print('1. them lien lac :') 
    print('2. xoa lien lac :')
    print('3. tim lien lac :')
    print('4. show all')
    print('5. exit')
    luachon = input('lua chon cua may la gi từ (1-5) :')
    if luachon == '1':
        # them lien lac moi vao dict
        ten = input('nhap ten :')
        sdt = input('nhap sdt :')
        lienlac[ten] = sdt
        # gán giá trị của key là ten trong dict lienlac bằng giá trị sdt
        print('da them lien lac')
    elif luachon == '2':
        # xoa lien lac khoi dict
        ten = input('nhap ten can xoa :')
        if ten in lienlac :
            del lienlac[ten] 
            print('da xoa lien lac')
        else:
            print('ko co ma xoa')
    elif luachon == '3':
        # tim kiem lien lac trong dict
        ten = input('nhap ten can tim: ')
        if ten in lienlac :
            print(f'{ten} : {lienlac[ten]}')
            # f' ' là f-string là Python sẽ nhét giá trị của biến (ở trong {}) vào trong chuỗi ngay khi in ra
        else:
            print('ko co ng can tim ')
    elif luachon == '4':
        # hien thi tat ca lien lac trong dict
        print('tat ca danh ba :')
        for n, p in lienlac.items():
            # n,p là bieensn tạm thời để lấy key và value của dict, ở đây n=key(ten), p=valua(sdt)
            print(f'{n} : {p}')
    elif luachon == '5':
        # exit
        print('thoat trg trinh')
        break
    else:
        print('lua chon ko hop le')


```
